# Supervised Learning: More Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

Have the students:
- practice implementing regression algorithms
- checking that the outputs make sense by plotting and calculating error

## Simple Linear Regression
- Provide a *small* example/data set (either made up or from kaggle or somewhere similar)
- Have them analyze the data using: least squares by hand (for 3-5 of the data points), using NumPy (total dataset), and sklearn's `LinearRegression`
- Ask them to calculate MAE and MSE for **each** of their models and discuss the differences/takeaways
- Some discussion about this can be found [here](https://stackoverflow.com/questions/36573046/difference-between-numpy-linalg-lstsq-and-sklearn-linear-model-linearregression)

## Multiple Linear Regression
- Write a problem that requires multiple linear regression
- Ask them to discuss why it requires multiple linear regression (i.e. justify using multiple vs. simple linear regression)
- Step them through how to implement the code
- Ask them to calculate MAE and MSE for their model
